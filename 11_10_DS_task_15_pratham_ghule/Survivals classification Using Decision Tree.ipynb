```python
#importing libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
df.drop(labels = {'passenger_id' , 'name' ,'boat' , 'home.dest','cabin','ticket','body'} , axis = 1 , inplace = True)
```

```python
# drops all the rows containing null values from the dataframe 
df = df.dropna()
```

```python
label_enc=LabelEncoder()
df["sex"]=label_enc.fit_transform(df["sex"])
df["embarked"]=label_enc.fit_transform(df["embarked"])
```

```python
df
```

```python
df.corr()
```

```python
# x table is independant variables and y is target
y = df[["survived"]]
x = df.drop(["survived"],axis=1)
```

```python
#making training and testing frames
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30,shuffle="true")
```

```python
#fitting Decision Tree Classifier on training dataset
model = DecisionTreeClassifier().fit(x_train,y_train)
```

```python
#predicting outcome
y_pred = model.predict(x_test)
```

```python
#Finding accuracy of model
accuracy_score(y_test,y_pred)
```
